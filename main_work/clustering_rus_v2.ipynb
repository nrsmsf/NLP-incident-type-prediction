```python
from pathlib import Path
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale

import sklearn
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import sklearn.metrics as sm
from sklearn.metrics import confusion_matrix, classification_report

import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.decomposition import TruncatedSVD
from scipy import sparse

import pickle
import operator
```

```python
ru_file = "./new_tokened_and_cleaned.csv"
```

```python
df = pd.read_csv(ru_file, dtype=str, sep=';')
```

```python
df.shape
```

```python
df['length'] = df['cleanedtext'].apply(lambda x: len(x))
cond = df['length'] > 40
df = df[cond]
df = df.drop('length', 1)

df.shape
```

```python
df.head()
```

```python
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1, 1), min_df=1)
#max_features=20000, ngram_range=(1, 2)

response = tfidf.fit_transform(df['cleanedtext'])

#sparse.save_npz("tfidf_csr_matrix.npz", response)
```

```python
n_clusters=20
km = KMeans(n_clusters=n_clusters, random_state=1, init='k-means++')
km.fit(response)

#labels = km.fit_predict(response)
print(" --------------------- ")
print("   Коэффициент силуэта : %0.3f"
        % sm.silhouette_score(response, km.labels_, sample_size=1000))
print(" --------------------- ")
```

```python

transformed = km.transform(response)
df_tr = pd.DataFrame(transformed, columns=
                     ['clus0',
                      'clus1',
                      'clus2',
                      'clus3',
                      'clus4',
                      'clus5',
                      'clus6',
                      'clus7',
                      'clus8',
                      'clus9',
                      'clus10',
                      'clus11',
                      'clus12',
                      'clus13',
                      'clus14',
                      'clus15',
                      'clus16',
                      'clus17',
                      'clus18',
                      'clus19'
                     ]
                    )

df_tr
df_tr.to_pickle('transformed_pickle.pkl')
#чем меньше расстояние, тем ближе к центру кластера -> строка принадлежит этому кластеру
```

```python
terms = tfidf.get_feature_names()
```

```python
svd = TruncatedSVD(2, random_state=10)
svdtransformed = svd.fit_transform(response)
svdtransformed.shape
```

```python
def display_topics(svd, terms, n_components, n_out = 7, n_weight = 5, topic = None):
 
    if topic is None:
        for k in range(n_components):
            idx = {i:abs(j) for i, j in enumerate(svd.components_[k])}
            sorted_idx = sorted(idx.items(), key=operator.itemgetter(1), reverse=True)
            weight = np.mean([ item[1] for item in sorted_idx[0:n_weight] ])
            print("T%s)" % k, end =' ')
            for item in sorted_idx[0:n_out-1]:
                print( " %0.3f*%s"  % (item[1] , terms[item[0]]) , end=' ')
            print()
    else:
        m = max(svd.components_[topic])
        idx = {i:abs(j) for i, j in enumerate(svd.components_[topic])}
        sorted_idx = sorted(idx.items(), key=operator.itemgetter(1), reverse=True)
        weight = np.mean([ item[1] for item in sorted_idx[0:n_weight] ])
        print("* T %s) weight: %0.2f" % (topic, weight), end=' ')
        for item in sorted_idx[0:n_out-1]:
            print( " %0.3f*%s"  % (item[1] , terms[item[0]]) , end=' ')
        print()

def plot_clusters(svdX, y_pred, centers):
    plt.style.use('fivethirtyeight')
    f, ax1 = plt.subplots(1, 1, figsize=( 16, 8), facecolor='white')
    ax1.set_xlabel("")
    ax1.set_ylabel("")
    ax1.set_title("K-Means")
    ax1.scatter(svdX[:,0], svdX[:,1], c=y_pred, cmap=plt.cm.Paired, s=45)
    ax1.axis('off')
    plt.show()
```

```python
display_topics(svd, terms,2)
```

```python
y_pred  = km.predict(response)
centers = km.cluster_centers_

plot_clusters(svdtransformed, y_pred, centers)
```
